```python
x = np.random.normal(0, scale =1, size=500)

g = sns.distplot(x, kde=False, fit=stats.norm);

(mu, sigma) = stats.norm.fit(x)

plt.ylabel('Frequency')

x_dummy = np.linspace(-10, 10, 500)
#g.plot(x_dummy, stats.norm.pdf(x_dummy, mu, sigma))
g.plot(x_dummy, stats.norm.pdf(x_dummy, mu+0.2, sigma+0.2))
g.plot(x_dummy, stats.norm.pdf(x_dummy, mu+0.4, sigma+0.4))
g.plot(x_dummy, stats.norm.pdf(x_dummy, mu+0.6, sigma+0.6))
g.plot(x_dummy, stats.norm.pdf(x_dummy, mu+0.8, sigma+0.8))
g.plot(x_dummy, stats.norm.pdf(x_dummy, mu-0.2, sigma+0.2))
g.plot(x_dummy, stats.norm.pdf(x_dummy, mu-0.4, sigma+0.4))
g.plot(x_dummy, stats.norm.pdf(x_dummy, mu-0.6, sigma+0.6))
g.plot(x_dummy, stats.norm.pdf(x_dummy, mu-0.8, sigma+0.8))

sns.plt.show()
```

```python
df1 = stats.norm.pdf(x_dummy, mu+0.2, sigma+0.2)
df2 = stats.norm.pdf(x_dummy, mu+0.4, sigma+0.4)
df3 = stats.norm.pdf(x_dummy, mu+0.6, sigma+0.6)
df4 = stats.norm.pdf(x_dummy, mu+0.8, sigma+0.8)
df5 = stats.norm.pdf(x_dummy, mu-0.2, sigma+0.2)
df6 = stats.norm.pdf(x_dummy, mu-0.4, sigma+0.4)
df7 = stats.norm.pdf(x_dummy, mu-0.6, sigma+0.6)
df8 = stats.norm.pdf(x_dummy, mu-0.8, sigma+0.8)
df = [df1,df2,df3,df4,df5,df6,df7,df8]
```

```python
def log_likelihood(x,mu=0,sigma=1):
    n = len(x)
    const = (-n/2)*np.log(2*np.pi*np.square(sigma))
    #return(const)
    tot = 0
    for i in x:
        s = np.square(i-mu)
        tot = tot+s
    var = (-0.5/np.square(sigma))*tot
    log_hood = const + var
    return log_hood
```

```python
log_df = []
for i in df:
    log_df.append(log_likelihood(i))
```
